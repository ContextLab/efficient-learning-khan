```python
import numpy as np
import pandas as pd
```

```python
raw = pd.read_csv('/Users/paxtonfitzpatrick/Desktop/thesis-test-df.txt', sep=';')
```

```python
df = raw[['Time', 'speech:transcription', 'character speaking:character speaking', 'scene details:description',
        'location:location', 'location:indoor/outdoor', 'shot:words onscreen', 'shot:camera angle', 'music present']]
```

```python
df = df.replace([-1000, str(-1000)], np.nan)
df['location:indoor/outdoor'] = df['location:indoor/outdoor'].replace(2.0, 'outdoor')
df['music present'] = df['music present'].replace({1.0: 'no', 2.0: 'yes'})
df['shot:words onscreen'] = df['shot:words onscreen'].replace(0.0, np.nan)
df['shot:camera angle'] = df['shot:camera angle'].replace({1.0: 'long', 2.0: 'medium', 3.0: 'close up'})
```

```python
df.columns = ['time','speech','character speaking','scene description','location','indoor/outdoor','text on screen',
             'camera angle','music present']
```
