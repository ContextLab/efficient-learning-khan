```python
import pandas as pd
```

```python
questions = pd.read_csv('eeg_questions.csv', header=None)
questions.columns = ['index', 'question', 'answer_1', 'answer_2', 'answer_3', 'answer_4']
questions = questions.replace(r'\n',' ', regex=True)
for col in ['question', 'answer_1', 'answer_2', 'answer_3', 'answer_4']:
    questions[col] = questions[col].apply(lambda x: x.strip())

questions
```

```python
qset1 = questions.iloc[:30, :]
qset2 = questions.iloc[30:60, :]
qset3 = questions.iloc[60:90, :]
```

```python
('{prompt: \'' + qset1['question'] + '\',\n options: [ \''+ qset1['answer_1'] + '\',\n\'' + qset1['answer_2'] + '\',\n\'' + qset1['answer_3'] + '\'], required: true}').to_csv('qset1.csv', index=False)
```
